```python
import numpy as np
import random
from data_aug.data_aug import *
from data_aug.bbox_util import *
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
import pickle as pkl
%matplotlib inline
```

```python
input_file = open("/home/dataset/DL_kaggle2/pascal_voc_training_data.txt", 'r')

#new bounding box text file
output_file = open("/home/yuan65536/DL-hw/kaggle02/test_augmented_data.txt", 'w')


# Data augmentation
for line in input_file:
    line = line.strip()
    line = line.split(" ")
    filename = line[0]
    obj_num = len(line)//5
    obj_list = []
    for i in range(obj_num):
        obj_list.append(line[(i*5+1):(i*5+6)])
        
    
    obj_np = (np.array(obj_list)).astype('int') 
    img = cv2.imread("/home/dataset/DL_kaggle2/VOCdevkit_train/VOC2007/JPEGImages/"+filename)[:,:,::-1]
    bboxes = obj_np.astype('float')
    print('file name',filename)
    seq = Sequence([RandomHSV(40, 40, 30),RandomHorizontalFlip(), RandomShear()])
    img_, bboxes_ = seq(img.copy(), bboxes.copy())
    cv2.imwrite("/home/yuan65536/DL-hw/kaggle02/augumented_test/"+"augmented_"+filename, img_)

    
# write new bounding box text file
    output_file.write("augmented_" + filename)
    bboxes_ = bboxes_.astype('int')
    for item in bboxes_:
        for detail in item:
            output_file.write(' ')
            output_file.write(str(detail))
            
    output_file.write("\n")
output_file.close()
```

```python
import os
import random
import shutil
from shutil import copyfile
from shutil import copy

#datasetpath
#size
def train_test_split(img_source_dir, train_size):
  
 
    
    #train_subdir = '/home/yuan65536/DL-hw/kaggle02/data/train'
    #validation_subdir = '/home/yuan65536/DL-hw/kaggle02/data/validation'
    

    train_counter = 0
    validation_counter = 0

    # ftxt
    f = open ( "/home/yuan65536/DL-hw/kaggle02/test_augmented_data.txt" , 'r')
    l = [ line.strip() for line in f ]
    no_samples = len(l)
    #filetxttrain
    #txt
    #file_val valtxt
    file = open('/home/yuan65536/DL-hw/kaggle02/data/train/pascal_voc_split_train_aug.txt','w')
    file_val = open('/home/yuan65536/DL-hw/kaggle02/data/valid/pascal_voc_split_val_aug.txt','w')
   
    data1 = []    
    data2 = []
    # Randomly assign an image to train or validation folder
    for filename in os.listdir(img_source_dir):

        if filename.endswith(".jpg") or filename.endswith(".png"): 
            fileparts = filename.split('.')
            filename_jpg = str(fileparts[0]) +'.jpg'
            
            #traindir
            if random.uniform(0, 1) <= train_size:
                copy(os.path.join(img_source_dir, filename), '/home/yuan65536/DL-hw/kaggle02/data/train')
                data1.append(filename_jpg)
                train_counter += 1
                
            #valdir
            else:
                copy(os.path.join(img_source_dir, filename), '/home/yuan65536/DL-hw/kaggle02/data/valid')
                data2.append(filename_jpg)
                validation_counter += 1
    
    for line in l: 
        for item in data1:
            if(line[0:10] == item or line[0:20] == item):
                file.write("%s\n" % line)
        for item in data2:
            if(line[0:10] == item or line[0:20] == item):
                file_val.write("%s\n" % line)
    file.close()
    file_val.close()
    print('Copied ' + str(train_counter) + ' images to data/train/')
    print('Copied ' + str(validation_counter) + ' images to data/valid/')
```
